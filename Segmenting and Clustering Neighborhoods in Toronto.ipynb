### Importing the necessary libraries

```python
import pandas as pd
import numpy as np
import requests
```

#### using pandas for reading the HTML file directly

```python
df_list=pd.read_html('https://en.wikipedia.org/wiki/List_of_postal_codes_of_Canada:_M')
df_list[0]
```

### converting the list to a data frame

```python
df=pd.DataFrame(df_list[0])
df
```

We need to process the cells that have an assigned borough. Ignore cells with a borough that is Not assigned.
We will use the where clause with appropriate filter

```python
filter=(df['Borough']!='Not assigned')
```

```python
df=df.where(filter,inplace=False)
df
```

```python
df=df.dropna()
df.head()
```
